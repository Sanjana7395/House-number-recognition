### Step 1 - Loading data

```python
import pandas as pd
import os
import numpy as np
from PIL import Image
import cv2
from sklearn.model_selection import train_test_split
import tensorflow as tf
import warnings
warnings.filterwarnings('ignore')
from tensorflow import keras
from tensorflow.keras.utils import to_categorical

import matplotlib.pyplot as plt
%matplotlib inline

file = pd.read_csv (r'/home/pavan/Downloads/2017-IWT4S-HDR_LP-dataset/trainVal.csv')
```

```python
y = file[['lp']].to_numpy()
url = file[['track_id']].to_numpy()

img = []
count = 0
for i in url:
    inter = np.array2string(i)[4:-2]
    full_url = os.path.join('/home/pavan/Downloads/2017-IWT4S-HDR_LP-dataset/', inter)
    img.append(cv2.imread(full_url))
    count += 1
```

### Step 2 - Splitting data into training and test dataset

```python
x_train,x_test,y_train,y_test=train_test_split(img,y,test_size=0.2)
```

### Step 3 - Vizualizing data

```python
def plot_images(img,labels,nrows,ncols,pred_labels=None):
    fig = plt.figure(figsize = (20,6));
    axes = fig.subplots(nrows,ncols)
    for i, ax in enumerate(axes.flat):
        ax.imshow(img[i])
        ax.set_xticks([]); ax.set_yticks([])
        if pred_labels is None:
            ax.set_title('True: %s' % labels.flatten()[i])
        else:
            ax.set_title('True: {0}, Pred: {1}'.format(labels[i], np.argmax(pred_labels[i])))
            
plot_images(x_train, y_train, 2, 8)
```

### Step 4 - Encapsulating each character within a rectangle using MSER detector

```python
from copy import deepcopy

def draw(rects, image, color):
    start = []
    end = []
    for r in rects:
        p1 = (r[0], r[1])
        p2 = (r[0]+r[2], r[1]+r[3])
        start.append(p1)
        end.append(p2)
        cv2.rectangle(image, p1,p2, color,1)
    return start, end

mser = cv2.MSER_create()
gray = cv2.cvtColor(img[0], cv2.COLOR_BGR2GRAY)
regions, _ = mser.detectRegions(gray)
hulls = [cv2.convexHull(p.reshape(-1, 1, 2)) for p in regions]
vis = deepcopy(img[0])

all_rects = list()
for i, contour in enumerate(hulls):
    x,y,w,h = cv2.boundingRect(contour)
    cv2.rectangle(vis, (x, y), (x+w, y+h), (0, 255, 0), 1)
    all_rects.append([x, y, w, h])

rects, weights = cv2.groupRectangles(all_rects, 1, 0.2)
filter_rect = []
for i in rects:
    if i[2]<i[3]:
        filter_rect.append(i)
print('Number of boxes: {}'.format(len(filter_rect)))
plt.imshow(vis)

vis2 = deepcopy(img[0])
start, end = draw(filter_rect, vis2, (0, 0, 255))

plt.imshow(vis2)
```

### Step 5 - Extracting each character

```python
sx = []
sy = []
for r in start:
    sx.append(r[1])
    sy.append(r[0])

ex = []
ey = []
for r in end:
    ex.append(r[1])
    ey.append(r[0])
digs = []

for i in range(7):
    digit = img[0][sx[i]:ex[i], sy[i]:ey[i]]
    digs.append(digit)

fig = plt.figure(figsize = (20,6));
axes = fig.subplots(1,7)    
for i, ax in enumerate(axes.flat):
    ax.imshow(digs[i])
    ax.set_xticks([]); ax.set_yticks([])
```
